```python
# Analysis no.2; To study the genre 
```

```python
# Checking the column of genre from the dataframe...
# By assigning the potential array in a genreArray variable...
```

```python
genreArray= anime_new_df[' genre'].unique()
genreArray
```

```python
for genres in genreArray:
    print(genres)
    print(len(genres))
```

```python
# importing AST modules to convert the representation above
# creating an empty dictionary to count each occurrence of the genre
```

```python
import ast

#creating an empty dict
genre_dict = {}

ind=0;
for genres in genreArray:
    genreList = ast.literal_eval(genres) #now we get "Action", "Adventure", etc... string array
    print("Length ",ind,": ", len(genreList))
    for genre in genreList:
        if genre in genre_dict:      # To check whether the dictionary has the item or not
            genre_dict[genre]+=1
        else:
            genre_dict[genre]=1       #if item is not in the dictionary, we assign count of 1 to initialize a new key for the dict
        print(genre)
    print("\n")
    ind+=1
```

```python
genre_dict
```

```python
# To find the exact number of Genres present in the anime dataset...
```

```python
print("There are ", len(genre_dict), " Genres of anime in this dataset")
```

```python
# To answer the question of highest number and lowest number according to genre...
```

```python
print("Highest number genre: ", max(genre_dict, key=genre_dict.get), "with count: ", genre_dict[max(genre_dict, key=genre_dict.get)])
print("Lowest number genre: ", min(genre_dict, key=genre_dict.get), "with count: ", genre_dict[min(genre_dict, key=genre_dict.get)])
```

```python
# Creating a pie chart using all the keys or items from the genre dictionary saved in a variable called pieLabels and values stored in a variable pieValues
```

```python
pieLabels=["%s" % key for key in genre_dict]
pieValues=["%d" % genre_dict[key] for key in genre_dict]
```

```python
# To show variable contents...
```

```python
pieLabels, pieValues
```

```python
figureObject, axesObject = plt.subplots(figsize=(40, 20))

box = axesObject.get_position()
axesObject.set_position([box.x0, box.y0, box.width * 1.3, box.height])


axesObject.pie(pieValues,labels=pieLabels,autopct='%1.2f', startangle=90, radius=200, frame=True)

# Aspect ratio - equal means pie is a circle

axesObject.axis('equal')

plt.title("Pie Chart of Anime Genre and its respective count")
plt.show()
```

```python
# Histogram...
# PS; Had to import numpy as np in order to define a numpy array of values for the line; y_pos = np.arange(len(pieLabels)) 

import numpy as np
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_axes([0,0,1,1])

y_pos = np.arange(len(pieLabels))

ax.barh(y_pos, pieValues, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(pieLabels)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Count')
ax.set_title('Genre Anime')

plt.show()
```
