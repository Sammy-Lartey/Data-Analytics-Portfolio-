##### Q1. Which book (title) recieved max reviews?

```python
max_reviews= df['n_reviews'].max()
review_book= df.loc[df['n_reviews'] == max_reviews, 'title'].iloc[0]
print(review_book, 'is the most reviewed book with a maximum number of ', max_reviews, 'reviews')
```

##### Visual Representation of Q1

```python
plt.figure(figsize=(15,15))
sns.barplot(x='n_reviews',y='title', data = df.head(10))
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.yticks(fontsize= 15)
plt.ylabel('Book Title')
plt.xlabel('Number of Reviews')
plt.title('Number of reviews per book')
```

##### Q2. Which book has the highest percentage of 5 stars?

```python
five_star = df['star5'].max()
five_star_book = df.loc[df['star5'] == five_star, 'title'].iloc[0]

print(five_star_book, ' had the highest percentage of 5 stars (', five_star, ')')
```

```python
# To confirm our theory, let's find out how many reviews 'Effective Data Science Infrastructure: How to make data scientists productive' had
```

```python
n_reviews_five_star_book = df.loc[df['title'] == five_star_book, 'n_reviews'].iloc[0]
print(five_star_book, 'had ',n_reviews_five_star_book, ' reviews out of which ', five_star, ' had 5 stars' )
```

```python
# let's try to check the 5 star percentages of books that had way more reviews than 'Effective Data Science Infrastructure: How to make data scientists productive' 
```

```python
maximum_reviews = df['n_reviews'].max()
five_star_on_max_n_reviews_book = df.loc[df['title'] == review_book, 'star5'].iloc[0]
print(review_book, 'had', five_star_on_max_n_reviews_book, 'percent of 5 stars and ', maximum_reviews, 'reviews')
```

##### Visual Representation of Q2

```python
plt.figure(figsize=(40,40))
sns.barplot(x='star5',y='title', data = df.head(20))
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.yticks(fontsize= 36)
plt.ylabel('Book Title')
plt.xlabel('Ratings')
plt.title('Books with 5 star ratings')
```

##### Q3. Which publisher (and author) is most famous?

```python
most_reviewed = df['n_reviews'].max()
star_publisher = df.loc[df['n_reviews'] == most_reviewed, 'publisher'].iloc[0]

print(star_publisher, ' had the maximum number of (', most_reviewed, ') reviews')
```

##### Visual Representation of Q3. a

```python
plt.figure(figsize=(40,18))
sns.catplot(x = 'n_reviews', y = 'publisher',data = df.head(50), kind = 'swarm')
plt.yticks(fontsize= 8)
plt.ylabel('Publisher')
plt.xlabel('number of reviews')
plt.title('Publishers and number of reviews per their books')
```

##### Author 

```python
most_reviewed = df['n_reviews'].max()
star_author = df.loc[df['n_reviews'] == most_reviewed, 'author'].iloc[0]

print(star_author, ' had the maximum number of (', most_reviewed, ') reviews')
```

```python
least_reviewed = df['n_reviews'].min()
_author = df.loc[df['n_reviews'] == least_reviewed, 'author'].iloc[0]

print(_author, ' had the least number of (', least_reviewed, ') reviews')
```

##### Visual Representation of Q3. b 

```python
plt.figure(figsize=(100, 100))
sns.catplot(x = 'n_reviews', y = 'author',data = df.head(20), kind = 'bar')
plt.yticks(fontsize= 12)
plt.ylabel('author')
plt.xlabel('number of reviews')
plt.title('Authors and number of reviews per their books')
```

##### Q5. What is the average number of pages in a data science book and what is the most expensive book?

```python
average_no_pages = df['pages'].mean()
print('Average number of pages per book is - ',average_no_pages)
```

##### Visual Representation of Q5. a

```python
plt.figure(figsize=(20, 20))
sns.catplot(x = 'pages', y = 'title',data = df.head(20), kind = 'bar')
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.yticks(fontsize= 6)
plt.ylabel('Title')
plt.xlabel('Pages')
plt.title('Title and number of pages')
```

##### What is the most expensive book?

```python
Highest_cost = df['price'].max()
most_expensive_book = df.loc[df['price'] == Highest_cost, 'title'].iloc[0]

print(most_expensive_book, 'is the most expensive book priced at $', Highest_cost)
```

##### What is the cheapest book?

```python
least_cost = df['price'].min()
Cheapest_book = df.loc[df['price'] == least_cost, 'title'].iloc[0]

print(Cheapest_book, 'is the cheapest book priced at $', least_cost)
```

##### Visual Representation of Q5

```python
plt.figure(figsize=(15,8))
sns.catplot(x = 'price', y = 'title',data = df.head(10), kind = 'swarm')
plt.yticks(fontsize= 6)
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.ylabel('title')
plt.xlabel('price in $')
plt.title('Title and its price')
```
